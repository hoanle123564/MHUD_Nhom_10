```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('./magic04.data', header=None)
df.columns = [
    "fLength", "fWidth", "fSize", "fConc", "fConc1", 
    "fAsym", "fM3Long", "fM3Trans", "fAlpha", "fDist", "class"
]

df.head()
```

## Phân bố nhãn (class)

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='class')
plt.title("Phân bố nhãn (g vs h)")
plt.xlabel("Class")
plt.ylabel("Số lượng")
plt.show()
```

## Phân bố thuộc tính **fLength**

```python
plt.figure(figsize=(8,4))
sns.histplot(df['fLength'], kde=True, bins=30)
plt.title("Histogram của fLength")
plt.xlabel("fLength")
plt.ylabel("Số lượng")
plt.show()
```
